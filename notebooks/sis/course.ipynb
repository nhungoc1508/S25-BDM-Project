Courses taken:
- Assuming every semester a student takes 5 classes on average
- One year: 10 classes (+- some noise)
- Assuming current time = spring semester
- Currently taking: 5
- Taken: 5 * n (semesters finished) +- noise
    - 1st year: n = 1
    - 2nd year: n = 3
    - 3rd year: n = 5
    - 4th year: n = 7

Grade cutoffs:
- \[90 - 100\]: A
- [80 - 90): B+
- [70 - 80): B
- [60 - 70): C+
- [50 - 60): C
- < 50: F

Grade distribution:
- A: 10%
- B+: 15%
- B: 25%
- C+: 25%
- C: 15%
- F: 10%

```python
grade_A = ['A'] * 10
grade_Bp = ['B+'] * 15
grade_B = ['B'] * 25
grade_Cp = ['C+'] * 25
grade_C = ['C'] * 15
grade_F = ['F'] * 10
g = {'grade': grade_A + grade_Bp + grade_B + grade_Cp + grade_C + grade_F}
tmp_df = pd.DataFrame(data=g)
grade_count = tmp_df.value_counts(normalize=True)
grade_count
```

```python
grade_count.index
```

```python
students['overall'] = np.random.choice(grade_count.index, size=len(students), p=grade_count.values)
students['overall'] = students.apply(lambda x: x['overall'][0], axis=1)
students.head()
```

```python
major_courses_dict = dict()
for dept in list(courses['dept_code'].unique()):
    major_courses_dict[dept] = courses[courses['dept_code'] == dept]['course_code'].to_list()
```

```python
n_dict = {1: 1, 2: 3, 3: 5, 4: 7}
cutoffs_dict = {
    'A': (90, 100),
    'B+': (80, 89),
    'B': (70, 79),
    'C+': (60, 69),
    'C': (50, 59),
    'F': (20, 49)
}

all_courses_codes = courses['course_code'].to_list()

def assign_courses(student):
    sems = n_dict[student['year']]
    cutoff = cutoffs_dict[student['overall']]
    dept = student['department']
    student_obj = dict()
    all_courses_taken = set()
    major_courses_codes = major_courses_dict[dept]
    # major_courses_codes = major_courses['course_code'].to_list()

    # Assign courses previously taken
    for semester in range(sems):
        student_courses = dict()
        num_course = 5 + random.randint(-1, 1)
        num_course_major = round(num_course * 0.6)
        num_course_misc = num_course - num_course_major
        # print(f'Sem: {semester}, all: {num_course}, major: {num_course_major}, misc: {num_course_misc}')
        for _ in range(num_course_major):
            course = random.choice(major_courses_codes)
            if (len(major_courses_codes) > 30):
                while course in all_courses_taken:
                    course = random.choice(major_courses_codes)
            student_courses[course] = random.randint(cutoff[0], cutoff[1])
            all_courses_taken.add(course)
        for _ in range(num_course_misc):
            course = random.choice(all_courses_codes)
            while course in all_courses_taken:
                course = random.choice(all_courses_codes)
            student_courses[course] = random.randint(cutoff[0], cutoff[1])
            all_courses_taken.add(course)
        student_obj[f'semester_{semester+1}'] = student_courses
    
    # Assign courses currently taking
    student_courses = dict()
    num_course = 5 + random.randint(-1, 1)
    num_course_major = round(num_course * 0.6)
    num_course_misc = num_course - num_course_major
    # print(f'Sem: {sems+1}, all: {num_course}, major: {num_course_major}, misc: {num_course_misc}')
    for _ in range(num_course_major):
        course = random.choice(major_courses_codes)
        if (len(major_courses_codes) > 30):
            while course in all_courses_taken:
                course = random.choice(major_courses_codes)
            student_courses[course] = None
            all_courses_taken.add(course)
    for _ in range(num_course_misc):
        course = random.choice(all_courses_codes)
        while course in all_courses_taken:
            course = random.choice(all_courses_codes)
        student_courses[course] = None
        all_courses_taken.add(course)
    student_obj[f'semester_{sems+1}'] = student_courses
    
    return student_obj
```

```python
t = time.time()
tmp = assign_courses(students.iloc[0])
print(time.time() - t)
tmp
```

```python
enrollment = []
for _, student in students.iterrows():
    id = student['student_id']
    student_obj = assign_courses(student)
    for sem, courses_obj in student_obj.items():
        for course, grade in courses_obj.items():
            is_taking = 1 if grade == None else 0
            row = {'student_id': id, 'course_code': course, 'semester': sem, 'is_taking': is_taking, 'grade': grade}
            enrollment.append(row)
```

```python
enrollment_df = pd.DataFrame.from_records(enrollment)
print(enrollment_df.shape)
enrollment_df.head()
```
