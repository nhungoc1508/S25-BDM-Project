```python
students = pd.read_csv(url_student)
print(students.shape)
students.head()
```

```python
students = students[['Student_Names', 'Phone_No.', 'Student Address']]
students.head()
```

```python
gender_detector = gender.Detector()
```

```python
names = students['Student_Names'].to_list()

names_clean = []
genders = []
student_ids = []
emails = []
id_dict = dict()

prefixes = ['Mr.', 'Ms.', 'Mrs.', 'Dr.', 'Miss', 'Mister', 'Sir']
suffixes = ['MD', 'PhD', 'DDS', 'DVM']
probs = 0
for name in names:
    # Clean name
    name_clean = name
    for prefix in prefixes:
        if prefix in name:
            name_clean = name.replace(f'{prefix} ', '')
    for suffix in suffixes:
        if suffix in name_clean:
            name_clean = name_clean.replace(f' {suffix}', '')
    if len(name_clean.split(' ')) > 2:
        name_split = name_clean.split(' ')
        name_clean = ' '.join(name_split[:2])
    names_clean.append(name_clean)

    # Get gender
    gender = gender_detector.get_gender(name_clean.split(' ')[0])
    if gender == 'mostly_male':
        gender = 'male'
    elif gender == 'mostly_female':
        gender = 'female'
    elif gender in ['andy', 'unknown']:
        gender = random.choice(['female', 'male'])
    genders.append(gender)

    # Generate student ID
    name_split = name_clean.split(' ')
    id = ''
    for name in name_split:
        id = id + str.lower(name[0])
    rand_num = random.randint(100, 999)
    if id not in id_dict:
        id_dict[id] = set([rand_num])
    while rand_num in id_dict[id]:
        rand_num = random.randint(100, 999)
    id_dict[id].add(rand_num)
    student_id = f'{id}{rand_num}'
    student_ids.append(student_id)

    # Generate email
    email = f'{student_id}@upc.edu'
    emails.append(email)
```

```python
for i in range(10):
    print(f'{names_clean[i], genders[i], student_ids[i], emails[i]}')
```

```python
students['Student_Names'] = names_clean
students['Gender'] = genders
students['ID'] = student_ids
students['email'] = emails

students.columns = ['name', 'phone', 'address', 'gender', 'student_id', 'email']
students.head()
```

Courses taken:
- Assuming every semester a student takes 5 classes on average
- One year: 10 classes (+- some noise)
- Assuming current time = spring semester
- Currently taking: 5
- Taken: 5 * n (semesters finished) +- noise
    - 1st year: n = 1
    - 2nd year: n = 3
    - 3rd year: n = 5
    - 4th year: n = 7

```python
def get_taking_num():
    return 5 + random.randint(-1, 1)

def get_taken_num(student):
    n_dict = {1: 1, 2: 3, 3: 5, 4: 7}
    n = n_dict[student['year']]
    return 5 * n + random.randint(-3, 3)
```

```python
students['courses_taking'] = students.apply(lambda x: get_taking_num(), axis=1)
students['courses_taken'] = students.apply(lambda x: get_taken_num(x), axis=1)
```
